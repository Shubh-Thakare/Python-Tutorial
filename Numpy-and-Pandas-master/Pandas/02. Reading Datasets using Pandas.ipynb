## Reading Datasets

**There are two multiple ways of getting datsets in python**

```We are Going to cover the most Popular type of Datasets that we deal with```

* Reading Delimited Files, where the data is separated using a delimiter.
* The Delimiter can be tab space, comma, etc 
* We can also read excel files using the function in pandas i.e., ```pd.read_excel()```
* We can read CSV Files using the function in pandas i.e., ```pd.read_csv()```
* We can read TSV Files i.e., tab separated variables using the same ```pd.read_csv('data.tsv', sep = '\t')``` mentioning the separator as tab.

````python
import pandas as pd
````

````python
# lets read the documentation of the read_csv function used to read csv files

help(pd.read_csv)
````

````python
# lets read a csv file using pandas

import pandas as pd
data = pd.read_csv('Datasets/employee.csv')
````
